```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

# Reading the Wine Data Set CSV file

```python
df = pd.read_csv("WineQT.csv")
```

# Using apply
I am using apply function to make the values of quality either 1 or 0 based on given threshold value.
<br>Threshold : <br>
0 : value<=6<br>
1 : value>6

```python
df['quality'] = df['quality'].apply(lambda x: 0 if x <= 6 else 1)
df.sample(5)
```

# Normalizing Other attributes by Min - Max Scaling

since we have to apply min-max scaling on all columns other than quality, hence we are dropping the column

```python
others = df.drop(columns=['quality'])
```

# without using Scikit learn class

```python
others2 = df.drop(columns=['quality'])
def scaledData(feature):
    min_val = feature.min()
    max_val = feature.max()
    return (feature - min_val) / (max_val - min_val)

# Running a loop to apply normalization to each feature
for feature in others2.columns:
    others2[feature] = scaledData(others2[feature])
```

```python
others2['quality'] = df['quality']
others2.sample(5)
```

# With using scikit learn class

```python
minMaxScalar = MinMaxScaler()

# Fitting and transforming the data
nArray = minMaxScalar.fit_transform(others)

# Since, nArray is an np-array, we have to convert it to dataFrame
dfNew = pd.DataFrame(nArray, columns=others.columns)

# Add the quality column back to dfNew
dfNew['quality'] = df['quality']
```

```python
# Sample of new data frame
dfNew.sample(5)
```
